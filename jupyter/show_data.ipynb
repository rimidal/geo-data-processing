```python
from pyproj import Proj, transform
from osgeo import ogr,osr
import json
```

```python
points_file1 = 'points500.txt'
points_file2 = 'points1000.txt'
poly_file1 = 'poly10.txt'
poly_file2 = 'poly15.txt'
f_500 = open('../TrainingDataSet/'+points_file1,'r')
f_1000 = open('../TrainingDataSet/'+points_file2,'r')
```

```python
points  = f_500.readlines()
points = [ '<'+x.split(':<')[1].replace('\n','') for x in points]
points = [ ogr.CreateGeometryFromGML(x) for x in points]
```

```python
points = [json.loads(x.ExportToJson())['coordinates'] for x  in points]
```

```python
points = pd.DataFrame(points)
```

```python
f_1000 = open('../TrainingDataSet/'+poly_file2,'r')
poly  = f_1000.readlines()
poly = [ '<'+x.split(':<')[1].replace('\n','') for x in poly]
poly = [ ogr.CreateGeometryFromGML(x) for x in poly]
```

```python
poly = [json.loads(x.ExportToJson())['coordinates'] for x  in poly]
```

```python
from bokeh.palettes import viridis
```

```python
poly = np.array([np.array(x).reshape(-1,2)for x in poly])
```

```python

lng_p = points[0].to_list()[0:9000]
lat_p = points[1].to_list()[0:9000]


v1 = np.random.rand(5)
v2 = np.random.rand(5)
# create a new plot with figure
p = figure(plot_width=900, plot_height=700)

for i in range(len(poly)):
    lng = poly[i][:,0]
    lat =poly[i][:,1]
    p.line(lat, lng,  line_color=viridis(len(poly))[i],  line_width=2)
p.circle(lat_p, lng_p,fill_color="red", size=2)

show(p) # show the results
```
